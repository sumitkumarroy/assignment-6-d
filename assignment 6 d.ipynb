Q1. What is Flask Framework? What are the advantages of Flask Framework?
Flask is a lightweight and versatile web application framework for Python. It is designed to be simple and easy to use, yet powerful enough to create a wide range of web applications, from small projects to large-scale applications.

Advantages of Flask:

Simplicity: Flask has a simple and easy-to-understand syntax, making it great for beginners.
Flexibility: Flask allows developers to choose the tools and libraries they want to use, rather than forcing a specific structure.
Lightweight: Flask is lightweight and has minimal dependencies, making it fast and efficient.
Extensibility: Flask has a modular design, allowing developers to add or remove components as needed.
Widely Adopted: Flask has a large community of developers and extensive documentation, making it easy to find help and resources.

```python
'''Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.'''
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run()

```

```python
'''Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the 
following details:'''

from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return '''
    <h1>Company Name: ABC Corporation</h1>
    <p>Location: India</p>
    <p>Contact Detail: 999-999-9999</p>
    '''

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

if __name__ == '__main__':
    app.run()
```

```python
'''Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for()function.
The url_for() function in Flask is used for URL building. It generates a URL to a specific function. This is particularly
useful when you want to avoid hardcoding URLs in your templates.'''

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    # Using url_for to generate URLs for other routes
    welcome_url = url_for('welcome')
    return f'<a href="{welcome_url}">Go to Welcome Page</a>'

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

if __name__ == '__main__':
    app.run()
```
